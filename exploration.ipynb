```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Nutrition-Summary-2021-11-01-to-2022-07-08.csv')
df.head()
```

```python

# df.to_csv('test.csv')
df = df.fillna(0)
df = df.drop(['Note'], axis=1)
# df = df.replace(np.nan, 0, inplace=True)
df.head()

```

```python

df = df.replace(0,df.median(axis=0))

df.head()
```

```python
df.head()
```

```python
print(df.mean())


```

```python
dfMacros = df[['Date', 'Fat (g)', 'Protein (g)', 'Carbohydrates (g)']]
dfMacros.head()
```

```python
#Visualize data
dfMacros2 = dfMacros[['Fat (g)', 'Protein (g)', 'Carbohydrates (g)']]
objects = dfMacros2.columns
y_pos = np.arange(len(objects))
values = dfMacros.mean(axis=0)
print(values)

plt.bar(y_pos, values, align='center')
plt.xticks(y_pos, objects)
plt.ylabel('Average Value/Meal')
for i, val in enumerate(values.to_list()):
    plt.text(val, i, str(val))
plt.show()

```
